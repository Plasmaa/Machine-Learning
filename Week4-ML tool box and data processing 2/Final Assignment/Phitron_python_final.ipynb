```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

## Answer to Q1

```python
df = pd.read_csv('sample_data/final-employee-ds.csv')

training_hours = np.array(df['TrainingHours'], dtype=float)

mean = np.mean(training_hours)
std = np.std(training_hours)

print("Mean:", mean)
print("Standard Deviation:", std)
```

## Answer to Q2

```python
df['Salary'] = df['Salary'].fillna(df.groupby('Department')['Salary'].transform('median'))

filtered_employees = df[(df['ExperienceYears'] > 12) & (df['ProjectCount'] > 5)]

total_salary = filtered_employees['Salary'].sum()

print("Total Salary Sum:", total_salary)
```

## Answer to Q3

```python
filtered_df = df[
    (df['Department'].isin(['IT', 'Finance'])) &
    (df['Age'] >= 30) & (df['Age'] <= 45) &
    (df['PerformanceScore'] > 88)
]

sorted_df = filtered_df.sort_values(by='Salary', ascending=False)

final_output = sorted_df[['Name', 'Department', 'Salary', 'PerformanceScore']]

print(final_output)
```

## Answer to Q4

```python
df['SalaryPerHour'] = df['Salary'] / (df['WorkHoursPerWeek'] * 4)

top_5_employees = df.sort_values(by='SalaryPerHour', ascending=False).head(5)

final_table = top_5_employees[['Name', 'SalaryPerHour']]

print(final_table)
```

## Answer to Q5

```python
plt.figure(figsize=(10, 6))

sns.countplot(data=df, x='Department', hue='Gender', palette='viridis')

plt.title('Employee Count by Department and Gender', fontsize=16)
plt.xlabel('Department', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.legend(title='Gender')

plt.show()
```
