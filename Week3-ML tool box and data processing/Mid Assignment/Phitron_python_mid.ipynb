```python
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

## Ans to Q1

```python

print("--- Ages Array Attributes ---")
print("Data Type:", ages.dtype)
print("Dimension:", ages.ndim)
print("Shape:", ages.shape)
print("Size:", ages.size)

print("\n--- Salaries Array Attributes ---")
print("Data Type:", salaries.dtype)
print("Dimension:", salaries.ndim)
print("Shape:", salaries.shape)
print("Size:", salaries.size)
```

## Ans to Q2

```python
max_salary = np.max(salaries)
min_salary = np.min(salaries)

print(f"Highest Salary: {max_salary}")
print(f"Lowest Salary: {min_salary}")

avg_salary = np.mean(salaries)
avg_age = np.mean(ages)

print(f"Average Salary: {avg_salary}")
print(f"Average Age: {avg_age}")
```

## Ans to Q3

```python
ages_more_30 = ages[ages > 30]
print("Ages greater than 30:", ages_more_30)

count = np.count_nonzero(ages > 30)
print("Number of employees older than 30:", count)
```

## Ans to Q4

```python
updated_ages = ages + 5

print("Original Ages:", ages)
print("Updated Ages (Original + 5):", updated_ages)
```

## Ans to Q5

```python
total_expense = np.sum(salaries)
print("Total Salary Expense:", total_expense)

salary_difference = np.max(salaries) - np.min(salaries)
print("Difference (Max - Min Salary):", salary_difference)
```
